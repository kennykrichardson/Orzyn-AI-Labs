```python
REPOSITORY_QUERY = """
query($owner:String!, $name:String!) {

    repository(owner:$owner, name:$name) {

        name

        description

        stargazerCount

        forkCount

        url

        isPrivate

    }

}
"""
```

```python
repo_config = get_active_repository()

OWNER = repo_config.owner

REPOSITORY = repo_config.repository
```

```python
repository = client.execute(

    REPOSITORY_QUERY,

    {

        "owner":OWNER,

        "name":REPOSITORY

    }

)

pretty_json(repository)
```

```python
PAGINATION_QUERY = """
query(
    $owner:String!,
    $name:String!,
    $after:String
){

repository(owner:$owner,name:$name){

issues(

first:100,

after:$after

){

nodes{

number

title

}

pageInfo{

hasNextPage

endCursor

}

}

}

}
"""
```

```python
issues = list(

client.paginate(

    PAGINATION_QUERY,

    {

        "owner":OWNER,

        "name":REPOSITORY

    },

    ["repository","issues"]

)

)

len(issues)
```
